```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

### Data Collection and Preprocessing

```python
#loading the csv data into a Panda DataFrame
df=pd.read_csv('CarPrice_Assignment.csv')
```

### Splitting the Features and Target

```python
X=df.drop("price",axis=1)
Y=df["price"]
```

```python
print(X)
```

```python
print(Y)
```

### Encoding the given data

```python

label_encoder=LabelEncoder() 
for col in X.columns:
    if X[col].dtype=='object':
        X[col]=label_encoder.fit_transform(X[col])   
```

```python
X.head()
```

### Splitting the data into traing and test data

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=42)  
```

### Linear Regression

```python
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
y_pred=lr_model.predict(x_test) 
```

```python
#lr_model evaluation
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
rmse
```

### Decision Tree

```python
dt_model= DecisionTreeRegressor(max_depth=None)
dt_model.fit(x_train,y_train)
dt_pred=dt_model.predict(x_test) 
```

```python
#dt_model evaluation
dt_mse=mean_squared_error(y_test,dt_pred)  
dt_rmse=np.sqrt(dt_mse)  
dt_rmse
```

### Random Forest 

```python
rf_model=RandomForestRegressor(max_depth=None)
rf_model.fit(x_train,y_train)
rf_pred=rf_model.predict(x_test)
```

```python
#rf_model evaluation
rf_mse=mean_squared_error(y_test,rf_pred)
rf_rmse=np.sqrt(rf_mse)
rf_rmse
```

```python
model_ev = pd.DataFrame({'Model': ['Linear Regression','Random Forest',
                    'Decision Tree'], 'Accuracy': [rmse,
                    rf_rmse,dt_rmse]})
model_ev
```
